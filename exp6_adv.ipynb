```python
# Function to convert range or handle special cases
def convert_range_to_avg(value):
    try:
        # Check if the value is a string
        if isinstance(value, str):
            # Handle range values (e.g., "105-130")
            if '-' in value:
                low, high = value.split('-')
                return (float(low) + float(high)) / 2
            # Handle values like "Up to 120" by extracting the number
            elif 'Up to' in value:
                return float(value.split(' ')[-1])
            else:
                return float(value)
        else:
            # If value is already numeric, return as is
            return value
    except ValueError:
        # If conversion fails, return NaN or a default value
        return None  # or you could use 0 or np.nan

# List of columns that need conversion
columns_to_convert = ['Height (cm)', 'Weight (kg)', 'Lifespan (years)', 'Average Speed (km/h)', 'Gestation Period (days)']

# Apply the conversion function to each of the listed columns
for column in columns_to_convert:
    df[column] = df[column].apply(convert_range_to_avg)

# Display the updated DataFrame
print(df)
```

```python
import re
# Function to convert range or handle special cases in 'Offspring per Birth'
def convert_offspring(value):
    try:
        # Remove any non-numeric characters like brackets, text etc.
        cleaned_value = re.sub(r'[^0-9\-]', '', value)

        # Handle range values (e.g., "2-5")
        if '-' in cleaned_value:
            low, high = cleaned_value.split('-')
            return (float(low) + float(high)) / 2
        else:
            return float(cleaned_value)
    except ValueError:
        # If conversion fails, return NaN or a default value
        return None  # or np.nan

# Apply the conversion function to 'Offspring per Birth' column
df['Offspring per Birth'] = df['Offspring per Birth'].apply(convert_offspring)

# Display the updated DataFrame
print(df)
```
